```python
import seaborn as sns
from scipy.stats import pearsonr
import matplotlib.pylab as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
plt.rc('figure', figsize=(15, 10))
df = pd.read_csv("dataset2.csv")
#df=df[:, 0:len(df.columns)-1]
corr_matrix = df.corr()

print(corr_matrix["Resilience score"].sort_values(ascending=False))

# plot the heatmap
# corr_matrix.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)

ax = sns.heatmap(corr_matrix, linewidth=0.7)
plt.show()
```
